# Aktivierungs- und Fehlerfunktionen

## Imports

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## Datengrundlage

```python
vector = np.linspace(-10,10,10000)
```

## Aktivierungsfunktionen

### Heaviside-Funktion

```python
def step(x):
    return np.where(x<0,0,1)
```

```python
plt.figure(figsize=(8,6))
plt.plot(vector, step(vector),c="#1ACC94")
plt.xlim(-5,5)
plt.ylim(-1.5,2.5)
plt.vlines(0,-1.5,6,ls="--",alpha="0.4")
plt.hlines(0,-5,5,ls="--",alpha="0.4")
plt.title("Heaviside-Funktion (Schrittfunktion)")
plt.show()
```

### Sigmoidfunktion

```python
def sigmoid(x):
    return 1.0/(1 + np.exp(-x))
```

```python
plt.figure(figsize=(8,6))
plt.plot(vector, sigmoid(vector),c="#1ACC94")
plt.xlim(-5,5)
plt.ylim(-1.5,2.5)
plt.vlines(0,-1.5,6,ls="--",alpha="0.4")
plt.hlines(0,-5,5,ls="--",alpha="0.4")
plt.title("Sigmoidfunktion")
plt.show()
```

### Tangens Hyperbolicus Funktion

```python
def tanh(x):
    return 1-(2/(np.exp(2*x)+1))
```

```python
plt.figure(figsize=(8,6))
plt.plot(vector, tanh(vector),c="#CC1B58")
plt.xlim(-5,5)
plt.ylim(-1.5,2.5)
plt.vlines(0,-1.5,6,ls="--",alpha="0.4")
plt.hlines(0,-5,5,ls="--",alpha="0.4")
plt.title("Tangens Hyperbolicus Funktion")
plt.show()
```

## Fehlerfunktionen

```python
def logloss(y, pred):
    return -(y*(np.log(pred)) + (1-y)*np.log(1-pred))
```

```python
predictions = np.linspace(0.0001,0.9999)

plt.figure(figsize=(8,6))
plt.plot(predictions, logloss(np.ones(50),predictions),label="Fehler für $y=1$",c="#1ACC94")
plt.plot(predictions, logloss(np.zeros(50),predictions),label="Fehler für $y=0$",c="#CC1B58")
plt.xlabel("Prädiktion $\hat{y}$")
plt.ylabel("Fehler")
plt.title("Log Loss / Kreuzentropie")
plt.legend()
plt.show()
```
