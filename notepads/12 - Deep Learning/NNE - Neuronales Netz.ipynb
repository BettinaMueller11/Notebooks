## Imports

```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
```

## Datengrundlage

```python
data = load_iris()
x = data["data"][:,:2]
y = np.where(data["target"]>=1,1,0) # target binarisieren

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=45)
```

## Neuronales Netz

```python
class SimpleNeuralNetwork:
    def __init__(self):
        self.weights1 = np.array([[0.1,0.5],[0.3,0.7]])
        self.weights2 = np.array([-0.5,1.5])
        
    def sigmoid(self,z):
        return 1.0/(1 + np.exp(-z))
    
    def predict(self,x):
        logit_layer1 = np.sum(x*weights1,axis=1)
        activation_layer1 = self.sigmoid(logit_layer1)

        logit_layer2 = np.sum(activation_layer1*weights2)
        activation_layer2 = self.sigmoid(logit_layer2)
        return activation_layer2
```

```python
nn = SimpleNeuralNetwork()
for xi in x_train:
    nn.predict(xi)
```
